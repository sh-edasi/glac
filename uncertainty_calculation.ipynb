```python
import glacierml as gl
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = gl.load_notebook_data('4')
df['STD'] = np.sqrt(df['Model Variance'])
df
```

```python
df['VolSTD'] = df['Area'] * (df['STD'] / 1e3)

df['VolUnc'] = df['VolSTD'] / df['Weighted Volume (km3)'] * 100

glac_sum = np.round(sum(df['Weighted Volume (km3)']) / 1e3, 3)

glac_sum_unc = np.round(np.sqrt(
    sum(df['VolSTD']**2)
) / 1e3, 3)

perc = np.round(glac_sum_unc / glac_sum * 100, 3)

print(
    f'Sum thickness: {glac_sum} +/- {glac_sum_unc} * 10^3 km^3, '
    f'{perc}% uncertainty'
)
```

```python
glac_res = pd.read_csv('GlaThiDa_residual_stats.csv')
glac_res['STD'] = np.sqrt(glac_res['Model Variance'])
glac_res['VolUnc'] = glac_res['STD'] / 1e3 * glac_res['Area']
glac_res['PerThickUnc'] = glac_res['STD'] / glac_res['Thickness'] *100
glac_res['PercVolUnc'] = glac_res['VolUnc'] / (glac_res['Thickness'] / 1e3 * glac_res['Area']) * 100
glac_res
```
