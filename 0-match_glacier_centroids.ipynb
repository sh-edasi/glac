```python
import glacierml as gl
import os
import pandas as pd
import numpy as np
```

```python
RGI = gl.load_RGI()
```

```python
home_path = '/data/fast1/glacierml/data/'
glathida_data = 'T_data'
glathida_path = os.path.join(home_path,glathida_data)
```

```python
glathida = pd.read_csv(glathida_path + '/T.csv')
glathida = glathida.dropna(subset = 'MEAN_THICKNESS')
glathida = glathida.reset_index().drop('index',axis = 1)
```

```python
glathida = pd.concat(
    [
        glathida,
        pd.Series(np.nan,name = 'RGIId',index = glathida.index),
        pd.Series(np.nan,name = 'RGI Centroid Distance',index = glathida.index)
    ], axis = 1
)
```

```python
df = gl.match_GlaThiDa_RGI_index(RGI,glathida,verbose = True, useMP = True)
```
