```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize = (10,10))
ax.add_patch(plt.Circle((100, 100),10, color='r', alpha=0.5))
ax.add_patch(plt.Circle((103.3, 105.4), 10, color='g', alpha=0.5))
plt.scatter(
    100,100,marker = 'x',color = 'r'
)
plt.scatter(103.3,105.4,marker = 'x', color = 'g')
plt.plot(
    (100,110),
    (100,100),
    color = 'k',
    linestyle = '--'
)
plt.text(
    106.5,
    100.3,
    '$\hat{r}_R$',
    fontsize = 14
)
plt.plot(
    (100,103.3),
    (100,105.4),
    color = 'k',
    linestyle = '--'
)
plt.text(
    100.999999,
    102.7,
    '$x$',
    fontsize = 14
)

x = np.sqrt(2.3**2 + 5.4**2)
r = 10
D = x / r
print(x)
print(D)

plt.plot(
    (100,100 + 10*D),
    (100,100),
    color = 'r',
    linestyle = '--'
)
plt.text(
    103.5,
    100.3,
    '$D$',
    color = 'r',
    fontsize = 14
)



ax.add_patch(
    plt.Circle((100,100), 2.5, color='k',linestyle = '--', alpha=0.25,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 5, color='k',linestyle = '--', alpha=0.25,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 7.5, color='k',linestyle = '--', alpha=0.25,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 11.5, color='k',linestyle = '--', alpha=0.25,fill = False)
)
plt.text(
    100,96.7,'$T = 0.25$', fontsize = 12.5
)
plt.text(
    100,94.25,'$T = 0.5$', fontsize = 12.5
)
plt.text(
    100,91.7,'$T = 0.75$', fontsize = 12.5
)
plt.text(
    100,87.7,'$T = 999$', fontsize = 12.5
)

ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.
plt.xticks([]) 
plt.yticks([]) 
# plt.title('Random Glacier')
plt.show()
```

```python
import matplotlib.pyplot as plt
r1 = 7.599
r2 = 10
fig, ax = plt.subplots(figsize = (10,10))
ax.add_patch(
    plt.Circle((100, 100), r1, color='g', fill = False, alpha = .5, linewidth = 3)
)
ax.add_patch(
    plt.Circle((100, 100), r2, color='r', fill = False, alpha = .5, linewidth = 3)
)

plt.plot(
    (100,100+r2),
    (100,100),
    color = 'r',
    linestyle = '--'
)
plt.text(
    108.5,
    100.3,
    '$\hat{r}_R$',
    fontsize = 14,
    color = 'r'
)
plt.plot(
    (100,100+r1),
    (100,100),
    color = 'g',
    linestyle = '--'
)
plt.text(
    101.5,
    100.3,
    '$\hat{r}_G$',
    fontsize = 14,
    color = 'g'
)



x = abs(r1**2-r2**2)/r2**2
print(x)

plt.plot(
    (100,100 - 10*x ),
    (100,100),
    color = 'k',
    linestyle = '--'
)
plt.text(
    98.5,
    100.3,
    '$A$',
    color = 'k',
    fontsize = 14
)




ax.add_patch(plt.Circle((100,100), 2.5, color='k',linestyle = '--', alpha=0.25,fill = False))
ax.add_patch(plt.Circle((100,100), 5, color='k',linestyle = '--', alpha=0.25,fill = False))
ax.add_patch(plt.Circle((100,100), 7.5, color='k',linestyle = '--', alpha=0.25,fill = False))
# ax.add_patch(plt.Circle((100,100), 5.8, color='purple',linestyle = '--', alpha=0.85,fill = False))

ax.add_patch(plt.Circle((100,100), 11.5, color='k',linestyle = '--', alpha=0.25,fill = False))

plt.text(
    100,96.7,'$T = 0.25$', fontsize = 12.5
)
plt.text(
    100,94.25,'$T = 0.5$', fontsize = 12.5
)
plt.text(
    100,91.7,'$T = 0.75$', fontsize = 12.5
)
plt.text(
    100,87.7,'$T = 999$', fontsize = 12.5
)
ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.
plt.xticks([]) 
plt.yticks([]) 
# plt.title('Random Glacier')
plt.show()
```
