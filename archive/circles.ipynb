```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize = (10,10))
ax.add_patch(plt.Circle((100, 100),10, color='#4B0092', alpha=0.5))
ax.add_patch(plt.Circle((103.3, 105.4), 10, color='#1AFF1A', alpha=0.5))

plt.plot(
    (100,110),
    (100,100),
    color = 'k',
    linestyle = '--'
)
plt.text(
    106.5,
    100.3,
    '$\hat{r}_R$',
    fontsize = 17
)
plt.plot(
    (100,103.3),
    (100,105.4),
    color = 'k',
    linestyle = '--',
)
plt.text(
    100.999999,
    102.7,
    '$x$',
    fontsize = 17
)

x = np.sqrt(2.3**2 + 5.4**2)
r = 10
D = x / r
print(x)
print(D)

plt.plot(
    (100,100 + 10*D),
    (100,100),
    color = 'r',
    linestyle = '--'
)
plt.text(
    103.5,
    100.3,
    '$D$',
    color = 'r',
    fontsize = 17
)



ax.add_patch(
    plt.Circle((100,100), 2.5, color='k',linestyle = '--', alpha=0.5,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 5, color='k',linestyle = '--', alpha=0.5,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 7.5, color='k',linestyle = '--', alpha=0.5,fill = False)
)
ax.add_patch(
    plt.Circle((100,100), 11.5, color='k',linestyle = '--', alpha=0.5,fill = False)
)
plt.text(
    98.75,96.7,'$T = 0.25$', fontsize = 17
)
plt.text(
    98.75,94.25,'$T = 0.5$', fontsize = 17
)
plt.text(
    98.75,91.7,'$T = 0.75$', fontsize = 17
)
plt.text(
    98.75,87.7,'$T = 999$', fontsize = 17
)
# plt.scatter(
#     100,100,marker = 'x',color = '#4B0092'
# )
plt.scatter(103.3,105.4,marker = 'x', color = '#1AFF1A')
ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.
plt.xticks([]) 
plt.yticks([]) 

lgd = ax.legend(loc='upper left',fontsize = 14)

def add_patch(legend):
    from matplotlib.patches import Patch
    ax = legend.axes

    handles, labels = ax.get_legend_handles_labels()
    handles.append(Patch(facecolor='#1AFF1A', edgecolor='k'))
    labels.append("GlaThiDa")
    handles.append(Patch(facecolor='#4B0092', edgecolor='k'))
    labels.append("RGI")

    legend._legend_box = None
    legend._init_legend_box(handles, labels)
    legend._set_loc(legend._loc)
    legend.set_title(legend.get_title().get_text())


add_patch(lgd)
plt.title('Coregistration Location Threshold', fontsize = 20)
plt.show()
```
